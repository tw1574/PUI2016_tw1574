## Map CitiBike Data to CT level

### Load CitiBike station location data
#### Next we need the Citibike station geographical data.  Fortunately for this exercise, it was already broken out in to the Station_id file and in our Data directory, so can be easily loaded.

```python
station_id=DataFrame.from_csv('Data/Station_id.csv')
#convert station_ids to string index
station_id.index=station_id.index.astype(str)
station_id.head()
```

### Create CitiBike station Geometry (Point) from Corrdinates (Lat & Lon)
#### Much like before with the CT shape file, we create the geometric areas for the Citibike stations

```python
import shapely
# combine lat and lon to one column
station_id['lonlat']=zip(station_id.lon,station_id.lat)
# Create Point Geometry for based on lonlat column
station_id['geometry']=station_id[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
station_id.head()
```

### Intersect the two geometries
#### This is bringing together the two shape file geometries - the CT shape file & Citibike stations

```python
a=[]
# check for each citibike stations to see what CT they are belong to
for i in range(len(station_id)):
    a.append(ct_shape[ct_shape.geometry.intersects(station_id.iloc[i].geometry)].BoroCT2010.values[0])
    print '\r',"%",str((i+1)*100./len(station_id))[:4],
station_id['BoroCT2010']=a
```

```python
station_id.BoroCT2010.head()
```

## Exploratory Geospatial Analysis
#### The pysal library will help us with the geospatial analysis

```python
import pysal as ps
```

### Spatial Weights
Contiguity: Queen Weights
#### To start with the geospatial analysis, we calculate the Queen Weights (i.e. the surrounding CTs from each other)

```python
qW_CT=ps.queen_from_shapefile(path)
```

#### Neighbors of index10
This shows us the surrounding CTs from the CT at index[10]

The cell below gives us an idea of the cardinality (# of distinct values)

```python
qW_CT[10]
```

```python
qW_CT.cardinalities[10]
```

#### Matrix of Weights
Builds a matrix of our Queen Weights

```python
Wmatrix, ids = qW_CT.full()
```

```python
Wmatrix
```

#### Change the matrix to be row-standardized

```python
qW_CT.transform = 'r'
```

### Spatial Lag  
Similarity of CT neibors can be measured by means of spatial lag as follows:
$$BikeRideCountLag = \sum_j w_{i,j} BikeRideCount$$

Using our previously created Monthly Avg and Queen Weight Matrix, we can calculate the Spatial Lag

```python
MonthlyRide=ct_shape_MonthlyAve_Zoom.MonthlyAve
MonthlyRideLag=ps.lag_spatial(qW_CT, MonthlyRide)
```

#### Create Quantiles for maping
##### This breaks up our ride Lags into deciles (q=10) of (approximately) similar sizes
Zero and less than zero skew the deciles a bit, but will not skew the results

```python
MonthlyRideLagQ10 = ps.Quantiles(MonthlyRideLag, k=10)
MonthlyRideLagQ10
```

#### Map the monthly rides lag

```python
f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom.assign(cl=MonthlyRideLagQ10.yb).plot(column='cl', categorical=True, 
        k=10, cmap='hot_r', linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("Monthly Ride Spatial Lag Deciles")
```

## Global  Spatial Autocorrelation

```python
b,a = np.polyfit(MonthlyRide, MonthlyRideLag, 1)
b,a 
```

### Moran Scatterplot

```python
f, ax = plt.subplots(1, figsize=(9, 9))

plt.plot(MonthlyRide, MonthlyRideLag, '.', color='firebrick')

 # dashed vert at mean of the last year's PCI
plt.vlines(MonthlyRide.mean(), MonthlyRideLag.min(), MonthlyRideLag.max(), linestyle='--', color="gray")
 # dashed horizontal at mean of lagged PCI
plt.hlines(MonthlyRideLag.mean(), MonthlyRide.min(), MonthlyRide.max(), linestyle='--', color="gray")

# red line of best fit using global I as slope
plt.plot(MonthlyRide, a + b*MonthlyRide, 'r')
plt.title('Moran Scatterplot')
plt.ylabel('Spatial Lag of MonthlyRides')
plt.xlabel('MonthlyRides')
```

### Moran's I

```python
I_MonthlyRide = ps.Moran(MonthlyRide, qW_CT)
```

```python
I_MonthlyRide.I, I_MonthlyRide.p_sim
```

Thus, the $I$ statistic is $0.656$ for this data, and has a very small $p$ value. 

```python
I_MonthlyRide.sim[0:10]
```

## Local  Spatial Autocorrelation

```python
LMo_MonthlyRide = ps.Moran_Local(MonthlyRide, qW_CT,permutations=999)
```

```python
LMo_MonthlyRide.Is[0:5], LMo_MonthlyRide.p_sim[0:5]
```

### LISA (Local Indicators of Spatial Autocorrelation)

```python
sigs = MonthlyRide[LMo_MonthlyRide.p_sim <= .001]
W_sigs = MonthlyRideLag[LMo_MonthlyRide.p_sim <= .001]
insigs = MonthlyRide[LMo_MonthlyRide.p_sim > .001]
W_insigs = MonthlyRideLag[LMo_MonthlyRide.p_sim > .001]
```

### Plot Signficant LISA

```python
f, ax = plt.subplots(figsize=(9, 9))
plt.plot(sigs, W_sigs, '.', color='firebrick')
plt.plot(insigs, W_insigs, '.k', alpha=.2)
plt.text(s='$I = %.3f$' % I_MonthlyRide.I, x=120000, y=5000, fontsize=18)

 # dashed vert at mean of the last year's PCI
plt.vlines(MonthlyRide.mean(), MonthlyRideLag.min(), MonthlyRideLag.max(), linestyle='--', color="gray")
 # dashed horizontal at mean of lagged PCI
plt.hlines(MonthlyRideLag.mean(), MonthlyRide.min(), MonthlyRide.max(), linestyle='--', color="gray")

# red line of best fit using global I as slope
plt.plot(MonthlyRide, a + b*MonthlyRide, 'r')
plt.title('Moran Scatterplot')
plt.ylabel('Spatial Lag of MonthlyRides')
plt.xlabel('MonthlyRides')
```

### Hot and Cold spots 

```python
sig = LMo_MonthlyRide.p_sim < 0.01
hotspots = LMo_MonthlyRide.q==1 * sig
coldspots = LMo_MonthlyRide.q==3 * sig
print "Number of Hotspots:", hotspots.sum()
print "Number of Coldspots:", coldspots.sum()
```

```python
from matplotlib import colors

hcmap = colors.ListedColormap(['grey','red','blue'])
hotcold = hotspots*1 + coldspots*2

f, ax = plt.subplots(figsize=(9, 9))
ct_shape_MonthlyAve_Zoom.assign(cl=hotcold).plot(column='cl', categorical=True, 
        k=10, cmap=hcmap, linewidth=0.1, ax=ax, 
        edgecolor='white', legend=True)
ax.set_axis_off()
plt.title("MonthlyAve Spatial Lag Deciles")
```
